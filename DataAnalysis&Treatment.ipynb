#Reporte de Análisis VIII

##Identificando y Excluyendo Outliers

```python
datos = pd.read_csv("Alquiler_residencial.csv", sep=";")
datos.head(10)
```

```python
datos.boxplot(["Valor"])
```

```python
datos[datos["Valor"] >= 500000]
```

```python
valor = datos["Valor"]
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
lim_inf = Q1 - 1.5 * IIQ
lim_sup = Q3 + 1.5 * IIQ
```

```python
seleccion = (valor >= lim_inf) & (valor <= lim_sup)
```

```python
datos_new = datos[seleccion]
```

```python
datos_new.boxplot(["Valor"])
```

```python
datos.hist(["Valor"])
```

```python
datos_new.hist(["Valor"])
```

```python
datos.head(10)
```

```python
datos.boxplot(["Valor"], by = ["Tipo"])
```

```python
grupo_tipo = datos.groupby("Tipo")["Valor"]
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
lim_inf = Q1 - 1.5 * IIQ
lim_sup = Q3 + 1.5 * IIQ
```

```python
datos_new2 = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos["Tipo"] == tipo
  eh_dentro_limite = (datos["Valor"] >= lim_inf[tipo]) & (datos["Valor"] <= lim_sup[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new2 = pd.concat([datos_new2,datos_seleccion])
```

```python
datos_new2.boxplot(["Valor"], by = ["Tipo"])
```

```python
datos_new2.to_csv("Alquiler_residencial_noOutliers.csv", sep=";", index=False)
```

```python
g1.scatter(datos_new2.Valor, datos_new2.Area)
g1.set_title("Valor x Area")

g2.hist(datos_new2.Valor)
g2.set_title("Histograma")

datos_g3 = datos_new2.Valor.sample(100)
datos_g3.index = range(datos_g3.shape[0])
g3.plot(datos_g3)
g3.set_title("Muestra (Valor)")

grupo4 = datos_new2.groupby("Tipo")["Valor"]
label = grupo4.mean().index
valores = grupo4.mean().values
g4.bar(label, valores)
g4.set_title("Valor Medio por Tipo")
```
